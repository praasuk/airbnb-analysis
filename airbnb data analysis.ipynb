### 📌 Why `histplot` is used instead of `barplot` for Outlier Detection

- `sns.histplot` is used to visualize the **distribution** of a continuous variable.
- It helps identify **outliers** by showing data spread, skewness, and long tails.
- `sns.barplot` shows **aggregated values** (like mean), which can hide outliers.
- Therefore, `histplot` is better suited for **univariate analysis** and **detecting unusual values**.


```python
#Price distribuion
df= data['price']

plt.figure(figsize=(8, 5))
sns.histplot(data=data, x='price', bins=100)
plt.title('Price Distribuition')
plt.ylabel("Frequency")

plt.xlim(0, 100000)
plt.xticks(np.arange(0, 100001, 5000), rotation=45)

plt.show()
```

.corr() is used to compute the correlation matrix between numeric columns, helping to identify relationships between variables.

A heatmap is an effective way to visually represent the correlation matrix, providing insights into the strength and direction of relationships between features in the dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlation matrix for selected columns
corr = df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()

# Set up the plot size
plt.figure(figsize=(8, 6))

# Create heatmap with a color palette and annotating values
sns.heatmap(data=corr, annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'label': 'Correlation Coefficient'}, vmin=-1, vmax=1)

# Title for the heatmap
plt.title('Correlation Heatmap of Selected Variables')

# Show the plot
plt.show()
```

Which listings are overpriced compared to others in the same neighborhood and room type?

```python
data.head(3)
```

```python
# Step 1: Calculate the average price for each (neighbourhood_group, room_type) combination
avg_prices = data.groupby(['neighbourhood_group', 'room_type'])['price'].mean().reset_index()
avg_prices.rename(columns={'price': 'avg_price'}, inplace=True)

# Step 2: Merge the average prices back to the original dataframe
merged = data.merge(avg_prices, on=['neighbourhood_group', 'room_type'])

# Step 3: Identify listings where price is significantly above average
# Define overpriced listings as those with price greater than 1.5 times the average
overpriced = merged[merged['price'] > 1.5 * merged['avg_price']]

# Step 4: Display the result
overpriced[['id', 'neighbourhood_group', 'room_type', 'price', 'avg_price']]
```

```python
w = data.groupby('neighbourhood_group')['rating'].mean().reset_index()
# Create a mapping from neighbourhood_group to avg rating
rating_map = dict(zip(w['neighbourhood_group'], w['rating']))

# Assign avg rating to each row in data based on neighbourhood_group
data['avg rating'] = data['neighbourhood_group'].map(rating_map)

# Filter listings with high avg rating and high listing count
more = data[data['avg rating'] > 4.5]
more_w = more[more['calculated_host_listings_count'] > 20]
more_w
```

```python
# Step 1: Select relevant columns
reliability_data = data[['host_name', 'number_of_reviews', 'reviews_per_month', 'availability_365']].copy()

# Step 2: Group by host to get aggregated values (if a host has multiple listings)
grouped = reliability_data.groupby('host_name').agg({
    'number_of_reviews': 'sum',
    'reviews_per_month': 'mean',
    'availability_365': 'mean'
}).reset_index()

# Step 3: Normalize each metric using Min-Max scaling
for col in ['number_of_reviews', 'reviews_per_month', 'availability_365']:
    grouped[f'norm_{col}'] = (grouped[col] - grouped[col].min()) / (grouped[col].max() - grouped[col].min())

# Step 4: Calculate reliability score (equal weight to all three)
grouped['reliability_score'] = (
    grouped['norm_number_of_reviews'] +
    grouped['norm_reviews_per_month'] +
    grouped['norm_availability_365']
) / 3

# Step 5: Rank hosts based on reliability score
ranked_hosts = grouped.sort_values('reliability_score', ascending=False)

# Display top 10 most reliable hosts
ranked_hosts[['host_name', 'reliability_score']].head(10)
```
